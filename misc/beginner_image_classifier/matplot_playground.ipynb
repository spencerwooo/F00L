```python
# Basic lines managing library import that you may need
import matplotlib.pyplot as plt
import numpy as np
```

## Simple plotting

Start with a simple plot:

```python
# Generate 100 evenly scattered value points between 0 and 2
x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')        # Plot a linear function: y = x
plt.plot(x, x**2, label='quadratic')  # Plot a quadratic function: y = x ^ 2
plt.plot(x, x**3, label='cubic')      # Plot a cubic function: y = x ^ 3

# Setting the x-axis and y-axis label
plt.xlabel('X values')
plt.ylabel('Y values')

# Setting the plot title
plt.title('Simple plot')

# Showing the plot legend
plt.legend()

# Showing the actual plot
plt.show()
```

Now we'll style our plot a little bit:

```python
# Evenly sampled time from 0. to 4. at 500ms intervals
t = np.arange(0., 4., 0.5)

# Plot y=x with color, marker and linestyle attributes
plt.plot(t, t, color='#ff9d46', marker='o', linestyle='--')
# Plot y=x^2 and y=x^3 with `[color][marker][line]` format strings
plt.plot(t, t**2, 'cs-.', t, t**3, 'm^')

# Draw plot
plt.show()
```

## Plotting with categorical variables

Let's plot some categorical variables, such as image prediction probabilities.

```python
# Defining some categories and the predictions our "model" came up with
categories = ['pineapple', 'avocado', 'mango']
predictions = [0.018, 0.951, 0.066]

# Whip up our figure (with three horizontal subplots)
# `figsize=(9,3)` means our figure is 9 inches wide and 3 inches tall
plt.figure(figsize=(9, 3))

# Plot 1: Bar plot, with a width of 0.35 (defaults to 0.8)
plt.subplot(131)
plt.bar(categories, predictions, width=0.35, color='#64b2cd')
# Plot 2: Scatter plot
plt.subplot(132)
plt.scatter(categories, predictions, color='#ffd739')
# Plot 3: Basic plot
plt.subplot(133)
plt.plot(categories, predictions, color='#6807f9')

# Define figure's super title
plt.suptitle('Categorical Variables Plotting')
plt.show()
```

## Dealing with multiple subplots

```python
# Function 1: y = e^(-t) * cos(2 *  * t)
def f1(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

# Function 2: y = cos(2 *  * t)
def f2(t):
    return np.cos(2 * np.pi * t)

# Sparsely scattered points to plot value markers
t1 = np.arange(0.0, 5.0, 0.2)
# Densely scattered points to smooth out the plot
t2 = np.arange(0.0, 5.0, 0.02)

# Define a 9 inches wide and 6 inches tall figure,
# with each image 9 inches wide and 3 inches tall
plt.figure(figsize=(9, 6))

# Subplot 1: Function 1's plot
plt.subplot(211)
plt.plot(t1, f1(t1), color='#293a80', marker='o', linestyle='')
plt.plot(t2, f1(t2), color='#293a80')

# Subplot 2: Function 2's plot
plt.subplot(212)
plt.plot(t2, f2(t2), color='#f39422', linestyle='--')
plt.show()
```

## Working with text

### Basic text insertion

We'll be plotting some random samples of the standard distribution model: $N(\mu, \sigma ^ 2)$.

```python
# N(mu, sigma^2): "standard normal" distribution
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Plot the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='#347474', alpha=0.9)

# Set labels
plt.xlabel('Smarts')
plt.ylabel('Probability')
# Set title
plt.title('Histogram of IQ')

# Set text showing the values of mu and sigma
plt.text(128, .027, r'$\mu=100,\ \sigma=15$')
# Define axis range of x and y
plt.axis([40, 160, 0.00, 0.03])
# Show grids
plt.grid(True)

plt.show()
```

### Annotations

```python
# x-axis data points, from 0.0 to 5.0, with pace of 0.01
t = np.arange(0.0, 5.0, 0.01)
# y-axis data points, calculated by function s = cos(2 *  * t)
s = np.cos(2 * np.pi * t)

# Plotting the actual figure
plt.plot(t, s, linewidth=2)

# Annotate the plot with arrow pointing at (2, 1) and text at (3, 1.5)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(width=1, headwidth=5, facecolor='black', shrink=0.1))

# y-axis data range from -2 to 2
plt.ylim(-2, 2)
plt.show()
```
